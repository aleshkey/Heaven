```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor

# Загрузка данных о пассажирах Титаника
data = pd.read_csv("..\\train.csv")
data.dropna(subset=['Age', 'Fare'], inplace=True)
# Предварительная обработка данных (например, заполнение пропущенных значений)

data
```

```python
# Извлечение столбцов "возраст" и "стоимость билетов"
ages = data["Age"].values
ticket_prices = data["Fare"].values

# Объединение столбцев в один массив признаков
X = pd.DataFrame({"Age": ages, "Fare": ticket_prices})

# Создание экземпляра модели LOF
lof = LocalOutlierFactor(n_neighbors=30)

# Обнаружение аномалий
anomaly_scores = lof.fit_predict(X)

# Вывод аномальных точек данных
anomalies = data[anomaly_scores == -1]
anomalies
```
