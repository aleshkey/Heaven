```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных для числа случаев заражения
confirmed_data = pd.read_csv('.\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_time_series\\time_series_covid19_confirmed_global.csv')
confirmed_data = confirmed_data.head(100)
# Загрузка данных для числа смертей
deaths_data = pd.read_csv('.\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_time_series\\time_series_covid19_deaths_global.csv')
deaths_data=deaths_data.head(100)
# Удаление ненужных столбцов (например, координаты)
confirmed_data = confirmed_data.drop(['Lat', 'Long', 'Province/State'], axis=1)
deaths_data = deaths_data.drop(['Lat', 'Long', 'Province/State'], axis=1)

# Группировка данных по странам/регионам
confirmed_data = confirmed_data.groupby('Country/Region').sum()
deaths_data = deaths_data.groupby('Country/Region').sum()

# Создание графика временного ряда для числа случаев заражения
confirmed_data.transpose().plot(figsize=(10, 6))
plt.xlabel('Дата')
plt.ylabel('Число случаев заражения')
plt.title('Изменение числа случаев заражения COVID-19 по времени')
plt.legend(loc='upper left')
plt.show()

# Создание графика временного ряда для числа смертей
deaths_data.transpose().plot(figsize=(10, 6))
plt.xlabel('Дата')
plt.ylabel('Число смертей')
plt.title('Изменение числа смертей от COVID-19 по времени')
plt.legend(loc='upper left')
plt.show()
```
