```python
import pandas as pd

# Загрузка набора данных из файла test.csv
titanic_data = pd.read_csv('..\\test.csv')

# Проверка наличия пропущенных значений
print("Количество пропущенных значений до обработки:")
print(titanic_data.isnull().sum())

# Удаление строк с пропущенными значениями
titanic_data_dropped = titanic_data.dropna()

# Заполнение пропущенных значений средними значениями (только для числовых столбцов)
titanic_data_filled_mean = titanic_data.fillna(titanic_data.select_dtypes(include='number').mean())

# Заполнение пропущенных значений медианами (только для числовых столбцов)
titanic_data_filled_median = titanic_data.fillna(titanic_data.select_dtypes(include='number').median())

# Проверка наличия пропущенных значений после обработки
print("\nКоличество пропущенных значений после обработки (удаление):")
print(titanic_data_dropped.isnull().sum())

print("\nКоличество пропущенных значений после обработки (заполнение средними значениями):")
print(titanic_data_filled_mean.isnull().sum())

print("\nКоличество пропущенных значений после обработки (заполнение медианами):")
print(titanic_data_filled_median.isnull().sum())
```
