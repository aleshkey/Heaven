```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

names = [
    '2015',
    '2016', 
    '2017', 
    '2018', 
    '2019'
]

datasets = []

for name in names:
    df = pd.read_csv(name+'.csv')
    datasets.append(df)

plt.figure(figsize=(20, 5))

# Визуализация столбчатой диаграммы для номинативной шкалы (страны)
plt.bar(datasets[0]['Country'], datasets[0]['Happiness Score'])
plt.xlabel('Страны')
plt.ylabel('Уровень счастья')
plt.title('Уровень счастья в разных странах')
plt.xticks(rotation=90)

plt.show()

# Визуализация scatter plot для показателей экономики и здоровья
plt.scatter(datasets[0]['Economy (GDP per Capita)'], datasets[0]['Health (Life Expectancy)'])
plt.xlabel('Экономика (ВВП на душу населения)')
plt.ylabel('Здоровье (Продолжительность жизни)')
plt.title('Связь между экономикой и здоровьем')
plt.show()
```

```python
plt.hist(datasets[1]['Happiness Score'], bins=10, edgecolor='black')
plt.xlabel('Уровень счастья')
plt.ylabel('Частота')
plt.title('Распределение уровня счастья')
plt.show()

# Визуализация box plot для показателей экономики и здоровья
plt.boxplot([datasets[1]['Economy (GDP per Capita)'], datasets[1]['Health (Life Expectancy)']])
plt.xticks([1, 2], ['Экономика', 'Здоровье'])
plt.ylabel('Значение')
plt.title('Сравнение экономики и здоровья')
plt.show()
```

```python

# Визуализация pair plot для сравнения нескольких показателей
sns.pairplot(datasets[2][['Happiness.Score', 'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom']])
plt.show()
print()
```
