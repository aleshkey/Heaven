```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

names = [
    '2015',
    '2016', 
    '2017', 
    '2018', 
    '2019'
]

datasets = []

for name in names:
    df = pd.read_csv(name+'.csv')
    datasets.append(df)

plt.figure(figsize=(20, 5))

#       ()
plt.bar(datasets[0]['Country'], datasets[0]['Happiness Score'])
plt.xlabel('')
plt.ylabel(' ')
plt.title('    ')
plt.xticks(rotation=90)

plt.show()

#  scatter plot     
plt.scatter(datasets[0]['Economy (GDP per Capita)'], datasets[0]['Health (Life Expectancy)'])
plt.xlabel(' (   )')
plt.ylabel(' ( )')
plt.title('    ')
plt.show()
```

```python
plt.hist(datasets[1]['Happiness Score'], bins=10, edgecolor='black')
plt.xlabel(' ')
plt.ylabel('')
plt.title('  ')
plt.show()

#  box plot     
plt.boxplot([datasets[1]['Economy (GDP per Capita)'], datasets[1]['Health (Life Expectancy)']])
plt.xticks([1, 2], ['', ''])
plt.ylabel('')
plt.title('   ')
plt.show()
```

```python

#  pair plot    
sns.pairplot(datasets[2][['Happiness.Score', 'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom']])
plt.show()
print()
```
