```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data = pd.read_csv('StudentsPerformance.csv')
print(data.columns)
```

```python
male_scores = data[data['gender'] == 'male']['math score']
female_scores = data[data['gender'] == 'female']['math score']

# Проведение t-теста
t_statistic, p_value = stats.ttest_ind(male_scores, female_scores)

# Вывод результатов
print('T-статистика:', t_statistic)
print('p-значение:', p_value)
```

```python
plt.figure(figsize=(10, 6))

# Гистограмма оценок по математике для мужчин
plt.subplot(1, 2, 1)
sns.histplot(male_scores, kde=True, color='blue')
plt.xlabel('Оценка по математике')
plt.ylabel('Частота')
plt.title('Распределение оценок по математике (мужчины)')

# Гистограмма оценок по математике для женщин
plt.subplot(1, 2, 2)
sns.histplot(female_scores, kde=True, color='pink')
plt.xlabel('Оценка по математике')
plt.ylabel('Частота')
plt.title('Распределение оценок по математике (женщины)')

plt.tight_layout()
plt.show()

# Визуализация результатов анализа
plt.figure(figsize=(6, 4))
sns.barplot(x=['Мужчины', 'Женщины'], y=[male_scores.mean(), female_scores.mean()])
plt.xlabel('Группа')
plt.ylabel('Средняя оценка')
plt.title('Сравнение средних оценок по математике')
plt.show()
```
