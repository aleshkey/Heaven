```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка набора данных о студентах (предположим, что у вас есть файл CSV с данными)
data = pd.read_csv('..\\test.csv')

# Вычисление описательных статистик
statistics = data.describe()

# Вычисление моды
modes = data.mode().iloc[0]

# Вывод описательных статистик и моды
print("Описательные статистики:")
print(statistics)
print("\nМоды:")
print(modes)

# Построение гистограмм
data.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```
