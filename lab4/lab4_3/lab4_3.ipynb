```python
import pandas as pd

# Загрузка данных в DataFrame
data = pd.read_csv('..\\test.csv')

# Применение one-hot encoding на столбце "sex"
encoded_sex = pd.get_dummies(data['Sex'], prefix='Sex')
data = pd.concat([data, encoded_sex], axis=1)

# Применение one-hot encoding на столбце "embarked"
encoded_embarked = pd.get_dummies(data['Embarked'], prefix='Embarked')
data = pd.concat([data, encoded_embarked], axis=1)

# Удаление исходных категориальных столбцов
data = data.drop(['Sex', 'Embarked'], axis=1)

# Вывод обновленного DataFrame
data
```
